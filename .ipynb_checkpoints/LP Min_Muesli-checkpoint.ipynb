```python
#circulating supply
import urllib.request
import pandas as pd
import json
# open a connection to a URL using urllib
webUrl  = urllib.request.urlopen('https://analytics.indigoprotocol.io/api/liquidity-pools/circulating-supply')

# read the data from the URL and print it
data = webUrl.read()
# print (data)

# byteData = b'Lets eat a \xf0\x9f\x8d\x95!'  
b1 = str(data, 'UTF-8')  
# print(b1)
type(b1)

json_object = json.loads(b1)
# print(type(json_object))

circulating_supply = pd.DataFrame(json_object)
circulating_supply=circulating_supply.drop(['id','timestamp', 'updated_at','asset'], axis=1)


circulating_supply.rename(columns = {'for':'For'}, inplace = True)
circulating_supply['DEX'] = circulating_supply.For.str.split(' ').str[0]
circulating_supply['iAsset'] = circulating_supply.For.str.split(' ').str[1]

ch = 'T'
# Remove all characters after the character '-' from string
# strValue = strValue.split(ch, 1)[0]

circulating_supply['created_at'] = circulating_supply.created_at.str.split(ch,1).str[0]
circulating_supply.rename(columns = {'created_at':'date'}, inplace = True)

circulating_supply.insert(2, column="actual_amount", value=circulating_supply['amount']/1000000)
# print(circulating_supply)

circulating_supply=circulating_supply.drop(['For'], axis=1)
# circulating_supply.insert(3, column="actual_amount", value=circulating_supply['amount']/1000000)
# # result = circulating_supply.loc[0, 'created_at']
# # print(result)
# # print(type(result))
# # circulating_supply
print(circulating_supply.to_string())
# list(circulating_supply.columns)
```

```python
import time
from datetime import date
import pandas as pd
from pycoingecko import CoinGeckoAPI

cg = CoinGeckoAPI()

# INDY
data1 = cg.get_coin_market_chart_by_id(id='indigo-dao-governance-token', vs_currency='usd', days=1000)
data1.pop('market_caps')
data1.pop('total_volumes')
dates1 = []
price1 = []
for i, j in data1["prices"]:
    new1 = str(pd.to_datetime(time.strftime("%d %b %Y", time.gmtime(i / 1000))))
    block = new1.split(" ")[0]
    if pd.to_datetime(block) == date.today():
        break
    dates1.append(block)
    price1.append(j)

# BTC
data2 = cg.get_coin_market_chart_by_id(id='bitcoin', vs_currency='usd', days=1000)
data2.pop('market_caps')
data2.pop('total_volumes')
dates2 = []
price2 = []
for i, j in data2["prices"]:
    new2 = str(pd.to_datetime(time.strftime("%d %b %Y", time.gmtime(i / 1000))))
    dates2.append(new2.split(" ")[0])
    price2.append(j)

# ETH
data3 = cg.get_coin_market_chart_by_id(id='ethereum', vs_currency='usd', days=1000)
data3.pop('market_caps')
data3.pop('total_volumes')
dates3 = []
price3 = []
for i, j in data3["prices"]:
    new3 = str(pd.to_datetime(time.strftime("%d %b %Y", time.gmtime(i / 1000))))
    dates3.append(new3.split(" ")[0])
    price3.append(j)

# iUSD
data4 = cg.get_coin_market_chart_by_id(id='tether', vs_currency='usd', days=999)
data4.pop('market_caps')
data4.pop('total_volumes')
dates4 = []
price4 = []
for i, j in data4["prices"]:
    new4 = str(pd.to_datetime(time.strftime("%d %b %Y", time.gmtime(i / 1000))))
    date = new4.split(" ")[0]
    if date in dates4:
        continue
    dates4.append(date)
    price4.append(j)

# Combine prices into a dictionary
dictionary_combo = {'Date': dates1, 'INDY_Price': price1, 'iBTC_Price': [], 'iETH_Price': [], 'iUSD_Price': []}
# Find common dates and corresponding prices
for i in range(len(dates1)):
    if dates1[i] in dates2:
        dictionary_combo['iBTC_Price'].append(price2[dates2.index(dates1[i])])
    if dates1[i] in dates3:
        dictionary_combo['iETH_Price'].append(price3[dates3.index(dates1[i])])
    if dates1[i] in dates4:
        dictionary_combo['iUSD_Price'].append(price4[dates4.index(dates1[i])])

df1 = pd.DataFrame(dictionary_combo)
print(df1.to_string())
```

```python
#epoch generator
def generate_arithmetic_progression(first_term, num_terms, common_difference):
    progression = []
    current_term = first_term
    
    for _ in range(num_terms):
        progression.append(current_term)
        current_term += common_difference
    
    return progression
first_term = 1685655900
num_terms = 29
common_difference = 86400

progression = generate_arithmetic_progression(first_term, num_terms, common_difference)
print(progression)
```
